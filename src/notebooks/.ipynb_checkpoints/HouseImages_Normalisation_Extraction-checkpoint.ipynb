```python
import cv2
import os
import sys
import csv
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
```

```python
#Reading the image
img = cv2.imread('Data/Household_air_study_Nov2019/LivingRoom_weekday/36hr_livingroom_weekday.jpg')
```

```python
#convert to RGB
imgRGB_original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#Showing the output
plt.imshow(imgRGB_original)
plt.grid()
```

```python
#Core portion extraction- Colour black
plt.imshow(imgRGB_original[2400:2600,2250:2450])
```

```python
#Finding the average colour values of the box 
R1 = imgRGB_original[2400:2600,2250:2450,0].mean()
G1 = imgRGB_original[2400:2600,2250:2450,1].mean()
B1 = imgRGB_original[2400:2600,2250:2450,2].mean()
```

```python
"""after having the mean values, identify the difference with original values 
from the colour card which is (52,52,52) for the colour black"""
```

```python
#Loop to apply the colour values changes to entire image
arr = np.array(imgRGB_original)
# grab the image dimensions
h1 = arr.shape[0]
w1 = arr.shape[1]  
# loop over the image, pixel by pixel
for y in range(0, h1):
    for x in range(0, w1):
        arr[y, x,0] = arr[y, x,0]+ (52- R1)  # add or substract the difference of Red
        arr[y, x,1] = arr[y, x,1]+ (52-G1) # add or substract the difference of green
        arr[y, x,2] = arr[y, x,2]+(52-B1) # add or substract the difference of Blue
```

```python
#Print the colour corrected image
plt.imshow(arr)
```

```python
#Saving back the normalised image
#output_filename = filename + '_Normalised.jpg'
plt.imsave('SampleImages/normalisedHouseImage2.jpg' ,arr)
```

```python
import cv2
import os
import sys
import glob
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image

path = "Data/Household_air_study_Nov2019/LivingRoom_weekday/*.jpg"
#path = "Data/Household_air_study_Nov2019/Kitchen_weekend/*.jpg"
#path = "Data/Household_air_study_Nov2019/Kitchen_weekday/*.jpg"

for file in glob.glob(path):
    filename = file.split('/')[3].split('.')[0]
    img = cv2.imread(file)
    imgRGB_original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    #Finding the average colour values of the core portion 
    R1 = imgRGB_original[2400:2600,2250:2450,0].mean()
    G1 = imgRGB_original[2400:2600,2250:2450,1].mean()
    B1 = imgRGB_original[2400:2600,2250:2450,2].mean()
    
    #Adjusting the colours in the image
    arr = np.array(imgRGB_original)
    # grab the image dimensions
    h1 = arr.shape[0]
    w1 = arr.shape[1]  
    # loop over the image, pixel by pixel
    for y in range(0, h1):
        for x in range(0, w1):
            arr[y, x,0] = arr[y, x,0]+ (52- R1)  # add or substract the difference of Red
            arr[y, x,1] = arr[y, x,1]+ (52-G1) # add or substract the difference of green
            arr[y, x,2] = arr[y, x,2]+(52-B1) # add or substract the difference of Blue
     
    #Saving the Normalised Images
    output_filename = filename + '_normalised.jpg'
    plt.imsave('Data/Household_air_study_Nov2019/LivingRoom_weekday/Normalised/'+output_filename ,arr)
    #plt.imsave('Data/Household_air_study_Nov2019/Kitchen_weekend/Normalised/'+output_filename ,arr)
    #plt.imsave('Data/Household_air_study_Nov2019/Kitchen_weekday/Normalised/'+output_filename ,arr)
```

```python
path = "Data/Household_air_study_Nov2019/LivingRoom_weekday/Normalised/*.jpg"
#path = "Data/Household_air_study_Nov2019/Kitchen_weekend/Normalised/*.jpg"
#path = "Data/Household_air_study_Nov2019/Kitchen_weekday/Normalised/*.jpg"

with open("Data/Household_air_study_Nov2019/LivingRoom_weekday/Normalised/ExtractedValues.csv", "w", newline="") as f:
#with open("Data/Household_air_study_Nov2019/Kitchen_weekend/Normalised/ExtractedValues.csv", "w", newline="") as f:
#with open("Data/Household_air_study_Nov2019/Kitchen_weekday/Normalised/ExtractedValues.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Image Name","Dot Number", "R Value", "G Value", "B Value"])
for file in glob.glob(path):
    filename = file.split('/')[4].split('_normalised')[0]
    img = cv2.imread(file)
    imgRGB_original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    #Initialization of the values
    dotNames = ['one','two','three','four','five','six','seven','eight','nine','ten','eleven','twelve','thirteen','fourteen','fifteen','sixteen']
    corX= [1235,1535,1835,2135]
    corY= [2800,3100,3400,3700]
    finalValues = []
    count = 0

    #Extracting the 5*5 square pixel mean colour values of each dot
    for dotx in corX:
        for doty in corY:
            valueR = imgRGB_original[dotx:dotx+5,doty:doty+5,0].mean()
            valueG = imgRGB_original[dotx:dotx+5,doty:doty+5,1].mean()
            valueB = imgRGB_original[dotx:dotx+5,doty:doty+5,2].mean()
            x = [filename,dotNames[count],valueR,valueG,valueB]
            finalValues.append(x)
            count= count+1

    #Saving the values into a csv file
    with open("Data/Household_air_study_Nov2019/LivingRoom_weekday/Normalised/ExtractedValues.csv", "a+", newline="") as f:
    #with open("Data/Household_air_study_Nov2019/Kitchen_weekend/Normalised/ExtractedValues.csv", "a+", newline="") as f:
    #with open("Data/Household_air_study_Nov2019/Kitchen_weekday/Normalised/ExtractedValues.csv", "a+", newline="") as f:
        writer = csv.writer(f)
        writer.writerows(finalValues)
        
```
