```python
"""after having the mean values, identify the difference woith  original values from the colour card"""
```

```python
#Loop to apply the colour values changes to entire image
arr = np.array(imgRGB_original)
# grab the image dimensions
h1 = arr.shape[0]
w1 = arr.shape[1]  
# loop over the image, pixel by pixel
for y in range(0, h1):
    for x in range(0, w1):
        arr[y, x,0] = arr[y, x,0]+ (52- R1)  # add or substract the applicable value
        arr[y, x,1] = arr[y, x,1]+ (52-G1) # add or substract the applicable value
        arr[y, x,2] = arr[y, x,2]+(52-B1) # add or substract the applicable value
```
